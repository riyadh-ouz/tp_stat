### Import libraries

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Théorème Central Limite

#### La population

```python
shape = 2.
scale = 2.
s = np.random.gamma(shape, scale, 1000000)

plt.hist(s)
plt.show()
```

#### Echantillonage

```python
mean_sample = [] # liste globale des moyennes

num_of_sample = [1000,2500,5000,10000,25000,50000]

sample_size = 500 # la taille d'échantillonage

for i in num_of_sample:
    # i représente le nombre d'échantillons
    each_mean_sample = []

    for j in range(0, i):

        rc = random.choices(s, k=sample_size)
        each_mean_sample.append(sum(rc)/len(rc))
        
    mean_sample.append(each_mean_sample)
```

#### Résultats et interprétation

```python
# Visualisation

cols = 2
rows = 3
fig, ax = plt.subplots(rows, cols, figsize=(20,15))
n = 0
for i in range(0, rows):
    for j in range(0, cols):
        ax[i, j].hist(mean_sample[n], 200, density=True)
        ax[i, j].set_title(label="number of samples :" + str(num_of_sample[n]))
        n += 1

plt.show()
```

```python
# Examinons le cas avec 50000 échantillons (le dernier cas)

sm = mean_sample[len(mean_sample) - 1]

# la déviation standard
std = np.std(sm)
# la moyenne
mean = np.mean(sm)

# Standardisation
zn = (sm - mean) / std
```

```python
# Comparaison avec la distribution normale standardisée

plt.figure(figsize=(20,10))
plt.hist(zn, 200, density=True)


mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),linewidth = 3, color='red')


plt.show()
```
